```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
labels = df['species'].copy()
df.species = label_encoder.fit_transform(df.species)
df.species.unique()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.69)
```

```python
train_x = train.iloc[:,0:4]
train_x
```

```python
train_y = train.species
train_y
```

```python
test_x = test.iloc[:,0:4]
test_y =test.species
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
tree_classifier = DecisionTreeClassifier(random_state=2)
tree_classifier.fit(train_x,train_y)
```

```python
y_pred = tree_classifier.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy score = {accuracy_score(test_y,y_pred)*100} %")
```
